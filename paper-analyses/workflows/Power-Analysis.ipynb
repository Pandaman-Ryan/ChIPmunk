# Power Analysis

```python
%pylab inline
import pandas as pd
import numpy as np
import os
import scipy.stats

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

# Paths
DATADIR="/storage/mlamkin/projects/tulip/paper-analyses/workflows/"
PDFDIR="/storage/mlamkin/projects/tulip/paper-analyses/workflows/"
FIG2COLOR = "darkblue"
```

## Fragment Mean Length

```python
fcolors = {
    "paired.hm": "darkblue",
    "paired.tf": "blue",
    "single.hm": "brown",
    "single.tf": "orange",
}

FMLDIR = DATADIR + "frag-mean-length/"

fig = plt.figure()
fig.set_size_inches((8, 6))
ax = fig.add_subplot(111)
for factor in fcolors.keys():
    data = pd.read_csv(os.path.join(FMLDIR, "summary_stats_fragmeanlen.%s.tab"%factor),
                sep="\t")
    data["fragmeanlen"] = [float(x[0])*float(x[1]) for x in data["fragmeanlen"].str.split('_')]

    if factor.split('.')[1] == 'tf':
        ax.plot((data["fragmeanlen"]), data["narrow-recall"], linewidth=0.75, 
                 marker="o", color=fcolors[factor], label=factor)
    else:
        ax.plot((data["fragmeanlen"]), data["broad-recall"], linewidth=0.75, 
                 marker="o", color=fcolors[factor], label=factor)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')
ax.axhline(y=1.00, linestyle="dashed", color="black", linewidth=0.75)
#ax.axhline(y=0.8, linestyle="dashed", color="black", linewidth=0.5)

ax.set_yticklabels(["%0.2f"%item for item in ax.get_yticks()], size=12)
ax.set_xlabel("Mean Fragment Length", size=15)
ax.set_ylabel("% Peaks recovered", size=15);
ax.legend(loc="lower right");
#fig.savefig(os.path.join(PDFDIR, "Fig2a_PowerAnalysis.pdf"))
```
