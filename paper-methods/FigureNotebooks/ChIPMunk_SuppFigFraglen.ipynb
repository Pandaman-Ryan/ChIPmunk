```python
%pylab inline

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

import pandas as pd
import os
import json
import scipy.stats as spst

PDFDIR="/storage/mgymrek/chipmunk_round2/pdfs"
```
